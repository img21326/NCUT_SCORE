```python
import pandas
master_course_imformation = pandas.read_csv('data/Outputs/imformation.csv')
master_course_innovation = pandas.read_csv('data/Outputs/innovation.csv')
master_course_other = pandas.read_csv('data/Outputs/other.csv')
master_course_technology = pandas.read_csv('data/Outputs/technology.csv')

master_course_imformation = master_course_imformation.drop(columns=['Unnamed: 0'])
master_course_innovation = master_course_innovation.drop(columns=['Unnamed: 0'])
master_course_other = master_course_other.drop(columns=['Unnamed: 0'])
master_course_technology = master_course_technology.drop(columns=['Unnamed: 0'])
master_course_other
```

```python
open_course = pandas.read_csv('data/Outputs/開課-日碩-上.csv')
# bachelor_open_course = bachelor_open_course.set_index('代碼')
open_course['english'] = False
open_course
```

```python
def match_course_func(df):
    match_course = []
    for master_index, master_row in df.iterrows():
        for open_index, open_row in open_course.iterrows():
            if ((type(master_row.科目) is str) and (type(open_row.科目) is str)):
                if (master_row.科目.strip() == open_row.科目.strip()):
                    if master_row.english:
                        open_row.english = True
                    match_course.append(open_row)
    match_course = pandas.DataFrame(match_course)
    return match_course
```

```python
match_course_func(master_course_imformation)
```

```python
match_course_func(master_course_innovation)
```

```python
match_course_func(master_course_other)
```

```python
match_course_func(master_course_technology)
```
