```python
import tabula
df = tabula.read_pdf('data/參考/daymaster_year_109.pdf')
```

```python
df = df[0]
df
```

```python
df_imformation = df[17:30]
df_imformation = df_imformation.drop('Unnamed: 1',axis=1)
df_imformation = df_imformation.drop('上學期 下學期',axis=1)
df_imformation = df_imformation.reset_index()
df_imformation = df_imformation.drop('index',axis=1)
df_imformation = df_imformation.rename(columns={'Unnamed: 0': '科目','Unnamed: 2': 'Subject'})
df_imformation['english'] = df_imformation['Subject'].apply(lambda x: 'True' if x[0] == "*" else 'False')
df_imformation['Subject'] = df_imformation['Subject'].apply(lambda x: x[1:] if x[0] == "*" else x)
df_imformation['科目'] = df_imformation['科目'].apply(lambda x: x[1:] if x[0] == "*" else x)

df_imformation
```

```python
def do(_from,_to):
    _df = df[_from:_to+1]
    _df = _df.drop('Unnamed: 1',axis=1)
    _df = _df.drop('上學期 下學期',axis=1)
    _df = _df.reset_index()
    _df = _df.drop('index',axis=1)
    _df = _df.rename(columns={'Unnamed: 0': '科目','Unnamed: 2': 'Subject'})
    _df['科目'] = _df['科目'].apply(lambda x: x[0:] if x[0] != "*" else x[1:])
    _df['Subject'] = _df['Subject'].apply(lambda x: x[0:] if x[0] != "*" else x[1:])
    _df['english'] = _df['Subject'].apply(lambda x: 'True' if x[0] == "*" else 'False')
    _df['Subject'] = _df['Subject'].apply(lambda x: x[1:] if x[0] == "*" else x)
    _df['科目'] = _df['科目'].apply(lambda x: x[1:] if x[0] == "*" else x)
    return _df
    
```

```python
df_mangement = do(17,29)
df_mangement
```

```python
df_technology = do(31,40)
df_technology
```

```python
df_innovation = do(42,48)
df_innovation
```

```python
df_other = do(50,52)
df_other
```

```python
arr_df = [
    df_imformation,
    df_technology,
    df_innovation,
    df_other
]
df_imformation.to_csv('data/Outputs/imformation.csv',encoding='utf-8-sig')
df_technology.to_csv('data/Outputs/technology.csv',encoding='utf-8-sig')
df_innovation.to_csv('data/Outputs/innovation.csv',encoding='utf-8-sig')
df_other.to_csv('data/Outputs/other.csv',encoding='utf-8-sig')
```
