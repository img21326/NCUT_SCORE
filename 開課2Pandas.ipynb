```python
import pandas
name = "全部開課.csv"
df = pandas.read_html('data/參考/全部開課-108-下.html',encoding='utf-8')
df = df[0]
df.head()
```

```python
df
```

```python
df = df.drop([6,7,9,10], axis=1)
df.head()
```

```python
# df['Subject'] = df[1].apply(lambda x: x.split(')')[1] if (type(x) is str and len(x.split(')')) > 0) else x)
df['count'] = df[1].apply(lambda x: len(x.split(')')) if (type(x) is str) else x)
df = df[df['count'] == 2]
```

```python
df['Subject'] = df[1].apply(lambda x: x.split(')')[1] if (type(x) is str) else x)
df['科目'] = df[1].apply(lambda x: x.split('(')[0]if type(x) is str else x)
df = df.drop([1], axis=1)
df
```

```python
df['時間'] = df[8].apply(lambda x: x.split('(')[0] if type(x) is str else x)
df['地點'] = df[8].apply(lambda x: x.split('(')[1][:-1] if type(x) is str else x)
df = df.drop([8], axis=1)
df
```

```python
df = df.rename(columns={0: '代碼',2: '學分',3:'學時',5:'班級',4:'老師'})
df = df.set_index('代碼')
df
```
